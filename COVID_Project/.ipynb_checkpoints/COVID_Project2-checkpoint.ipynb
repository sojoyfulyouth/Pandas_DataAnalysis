## 원하는 열 추려내기

### 열 항목 확인 및 수정된 데이터 프레임 생성
- iso_code : 국제 표준 국가 코드
- location : 국가 이름
- date : 일자
- total_cases : 전체 확진자 수
- population : 인구

```python
#원하는 열 list만들기 (df : data frame)
selected_columns = ['iso_code', 'location', 'date', 'total_cases', 'population']

#원하는 열을 추출한 데이터프레임 생성
revised_df = raw_df[selected_columns]

#수정된 데이터프레임 확인
revised_df.head()
```

### 유일한 데이터 추출

- Method : unique()

```python
# location 열 추출
location = raw_df['location']
location
```

```python
#location열의 유일한 값 확인
location.unique()
```

```python
kor_total_df = raw_df[raw_df.location=='South Korea']
kor_total_df.head()
```

```python
## South Korea 데이터 프레임 생성

kor_df = revised_df[revised_df.location=='South Korea']

kor_df.head()
```

```python
## United States 데이터 프레임 생성

usa_df = revised_df[revised_df.location=='United States']

usa_df.head()
```

### index 변경
- Method : set_index()

```python
# date로 index변경

kor_date_index_df = kor_df.set_index('date')
usa_date_index_df = usa_df.set_index('date')
```

```python
kor_date_index_df.head()
```

```python
usa_date_index_df.head()
```

## 확진자 비율 구하기

```python
kor_population = kor_date_index_df['population']['2020-01-22']

usa_population = usa_date_index_df['population']['2020-01-22']

print('미국 인구(', usa_population, ') / 대한민국 인구(', kor_population, ')')
```

```python
# 인구비율 구하기

rate = round((usa_population/kor_population), 2)
rate
```

```python
#확진자 비율을 적용한 최종 데이터 프레임 생성

final_revised_df = pd.DataFrame({
    'KOR' : kor_total_cases * rate,
    'USA' : usa_total_cases
    }, index=kor_date_index_df.index)

final_revised_df.head()
```

```python
final_revised_df[:].plot.line(rot=45)
```

```python
final_revised_df['2020-01-01':].plot.line(rot=45)
```

```python
usa_total_cases
```

```python
print(usa_population)
print(kor_population)
print(hawaii_population)
```

2. 캘리포니아, 텍사스, 버지니아, 일리노이, 워싱턴의 코로나 확진자 수 비율 변동 추이를 선 그래프로 그리고, 코로나 상황이 가장 완화된 주 알아보기

```python
raw_states_df = pd.read_csv('usa_main_states_covid.csv')
raw_states_df.head()
```

```python
filtered_states_df=raw_states_df[['state','submission_date', 'tot_cases']]
sorted_states_df=filtered_states_df.sort_values(by='submission_date')
sorted_states_df['date']=pd.to_datetime(filtered_states_df['submission_date'])
sorted_states_df.sort_values(by='date', inplace=True)
sorted_states_df.set_index('date', inplace=True)
sorted_states_df
```

```python
CA_total_df=sorted_states_df[sorted_states_df.state=='CA']
CA_population=39_029_342
CA_total_cases=CA_total_df['tot_cases']
CA_total_cases
```

```python
TX_total_df=sorted_states_df[sorted_states_df.state=='TX']
TX_population=30_029_572
TX_total_cases=TX_total_df['tot_cases']
TX_total_cases
```

```python
VA_total_df=sorted_states_df[sorted_states_df.state=='VA']
VA_population=8_683_619
VA_total_cases=VA_total_df['tot_cases']
VA_total_cases
```

```python
IL_total_df=sorted_states_df[sorted_states_df.state=='IL']
IL_population=12_582_032
IL_total_cases=IL_total_df['tot_cases']
IL_total_cases
```

```python
WA_total_df=sorted_states_df[sorted_states_df.state=='WA']
WA_population=7_864_400
WA_total_cases=WA_total_df['tot_cases']
WA_total_cases
```

```python
rate_CA = round((usa_population/CA_population), 2)
rate_TX = round((usa_population/TX_population), 2)
rate_IL = round((usa_population/IL_population), 2)
rate_VA = round((usa_population/VA_population), 2)
rate_WA = round((usa_population/WA_population), 2)

final_states_revised_df=pd.DataFrame({
    'CA':CA_total_cases*rate_CA,
    'TX':TX_total_cases*rate_TX,
    'IL':IL_total_cases*rate_IL,
    'VA':VA_total_cases*rate_VA,
    'WA':WA_total_cases*rate_WA,
    },index=sorted_states_df.index
)
final_states_revised_df
```

```python
final_states_revised_df.dropna(axis=0, inplace=True)
final_states_revised_df
```

```python
final_states_revised_df['2022-01-01':].plot(rot=45)
```

Result: 다섯 개 주의 확진자 수 변동 추이가 전반적으로 비슷함을 알 수 있다.
